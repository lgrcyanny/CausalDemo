```python
import causalnex
from causalnex.plots.plots import *
from IPython.display import Image
from causalnex.structure import StructureModel
import warnings
```

```python
sm_manual = StructureModel()
sm_manual.add_edges_from(
    [
        ("b", "a"),
        ("b", "c"),
        ("d", "a"),
        ("d", "c"),
        ("d", "b"),
        ("e", "c"),
        ("e", "b"),
    ],
    origin="expert",
)
```

```python
viz = plot_structure(sm_manual)
image_binary = viz.draw(format="jpg")
Image(image_binary)
```
